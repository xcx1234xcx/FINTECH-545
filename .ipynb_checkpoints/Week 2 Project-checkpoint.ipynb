```python
# Load the dataset
file_path = 'problem1.csv'
df = pd.read_csv(file_path)

data = df['x'].values
```

Q1

```python
# Define the function to calculate the first 4 moments using normalized formulas
def first4Moments(sample):
    n = len(sample)
    
    # Mean
    mu_hat = np.sum(sample) / n
    
    # Remove the mean from the sample
    sim_corrected = sample - mu_hat
    cm2 = np.dot(sim_corrected, sim_corrected) / n
    
    # Variance
    sigma2_hat = np.dot(sim_corrected, sim_corrected) / (n - 1)
    
    # Skewness
    skew_hat = np.sum(sim_corrected ** 3) / (n * np.sqrt(cm2 ** 3))
    
    # Kurtosis
    kurt_hat = np.sum(sim_corrected ** 4) / (n * cm2 ** 2)
    
    excessKurt_hat = kurt_hat - 3
    
    return mu_hat, sigma2_hat, skew_hat, excessKurt_hat

m, s2, sk, k = first4Moments(data)

print(f"Mean: {m}")
print(f"Variance: {s2}")
print(f"Skewness: {sk}")
print(f"Kurtosis: {k}")
```

Q3

```python
def test_biased(alpha = 0.05):
    samples = 1000
    sample_size = 100
    means = []
    variances = []
    skews = []
    kurts = []
    for _ in range(samples):
        x = np.random.normal(0, 1, sample_size)
        
        means.append(np.mean(x))
        variances.append(np.var(x))
        skews.append(skew(x))
        kurts.append(kurtosis(x))

    t_value_mean = np.mean(means) / (np.std(means) / np.sqrt(samples))
    t_value_variance = (np.mean(variances) - 1)/ (np.std(variances) / np.sqrt(samples))
    t_value_skewness = np.mean(skews) / (np.std(skews) / np.sqrt(samples))
    t_value_kurtosis = np.mean(kurts) / (np.std(kurts) / np.sqrt(samples))
    
    p_value_mean = 2 * (1 - t.cdf(abs(t_value_mean), samples-1))
    p_value_variance = 2 * (1 - t.cdf(abs(t_value_variance), samples-1))
    p_value_skewness = 2 * (1 - t.cdf(abs(t_value_skewness), samples-1))
    p_value_kurtosis = 2 * (1 - t.cdf(abs(t_value_kurtosis), samples-1))

    if p_value_mean > alpha and p_value_variance > alpha and p_value_skewness > alpha and p_value_kurtosis > alpha:
        print("We have no reason to reject the Null Hypothesis that Python package is not biased")
    else:
        print("We need to reject the hypothesis that Python package is not biased")
        
    print(p_value_mean, p_value_variance, p_value_skewness, p_value_kurtosis)
```

```python
test_biased(0.05)
```

# Problem 2

```python
# Load the dataset
file_path = 'problem2.csv'
df = pd.read_csv(file_path)

X = df['x'].values
y = df['y'].values

# Add a constant to the independent variables (for the intercept term)
X = sm.add_constant(X)
```

Q1

```python
# Fit the data using OLS
ols_model = sm.OLS(y, X).fit()

# Display OLS results
print("OLS Regression Results")
print(ols_model.summary())
```

```python
# Fit the data using MLE (normal distribution)

# Define the negative log-likelihood function
def negative_log_likelihood(beta):
    residuals = y - np.dot(X, beta)
    sigma = np.std(residuals, ddof=len(beta))
    n = len(y)
    nll = n * np.log(sigma) + np.sum(residuals ** 2) / (2 * sigma ** 2)
    return nll

# Initial beta values (OLS estimates)
initial_beta = np.array(ols_model.params)

# Use minimize from scipy to find MLE of the parameters
result = minimize(negative_log_likelihood, initial_beta)

# Extract MLE results
mle_beta = result.x
mle_sigma = np.std(y - np.dot(X, mle_beta), ddof=len(mle_beta))

print("\nMLE Results")
print(f"MLE Beta: {mle_beta}")
print(f"MLE Standard Deviation (Sigma): {mle_sigma}")
```

```python
# Compare OLS and MLE results
print("\nComparison of Results:")
print(f"OLS Beta: {ols_model.params}")
print(f"OLS Standard Deviation of Errors: {np.std(ols_model.resid, ddof=len(ols_model.params))}")
print(f"Difference in Beta: {mle_beta - ols_model.params}")
print(f"Difference in Standard Deviation (Sigma): {mle_sigma - np.std(ols_model.resid, ddof=len(ols_model.params))}")
```

Q2

```python
# Define the negative log-likelihood function for MLE with T-distribution
def negative_log_likelihood_t(params):
    beta = params[:-1]
    nu = params[-1]
    residuals = y - np.dot(X, beta)
    sigma = np.std(residuals, ddof=len(beta))
    n = len(y)
    nll = -np.sum(t.logpdf(residuals / sigma, df=nu)) + n * np.log(sigma)
    return nll

# Set initial values for beta (OLS estimates) and degrees of freedom (nu)
initial_params_t = np.append(ols_model.params, 5)  # Initial guess: nu = 5

# Use minimize from scipy to find MLE of the parameters assuming T-distribution
result_t = minimize(negative_log_likelihood_t, initial_params_t, bounds=[(None, None)] * len(initial_params_t[:-1]) + [(2.1, None)])

# Extract MLE results for T-distribution
mle_beta_t = result_t.x[:-1]
mle_nu_t = result_t.x[-1]
mle_sigma_t = np.std(y - np.dot(X, mle_beta_t), ddof=len(mle_beta_t))

# Display the results
print("MLE Results under T-distribution:")
print(f"MLE Beta (T-distribution): {mle_beta_t}")
print(f"MLE Degrees of Freedom (nu): {mle_nu_t}")
print(f"MLE Standard Deviation (Sigma) under T-distribution: {mle_sigma_t}")
```

```python
# Calculate adjusted R-squared for each model
def adjusted_r_squared(y, y_pred, p):
    n = len(y)
    r_squared = 1 - np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2)
    adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)
    return adj_r_squared
```

```python
# Predicted values
y_pred_mle = np.dot(X, mle_beta)
y_pred_mle_t = np.dot(X, mle_beta_t)

# Calculate adjusted R-squared for each model
adj_r_squared_mle = adjusted_r_squared(y, y_pred_mle, len(mle_beta))
adj_r_squared_mle_t = adjusted_r_squared(y, y_pred_mle_t, len(mle_beta_t))
```

```python
# Display the results
results = {
    'Model': ['MLE (Normal)', 'MLE (T-distribution)'],
    'Adjusted R-squared': [adj_r_squared_mle, adj_r_squared_mle_t]
}

print(results)
```

# Problem 3

```python
# Load the dataset
file_path = 'problem3.csv'
df = pd.read_csv(file_path)

x = df['x'].dropna()
```

```python
# Define a function to fit AR and MA models and compare AIC values
def fit_models(data):
    results = {}
    # Fit AR(1) to AR(3)
    for p in range(1, 4):
        model = ARIMA(data, order=(p, 0, 0))
        fit = model.fit()
        results[f'AR({p})'] = fit
        print(f'AR({p}) AIC: {fit.aic}')
    
    # Fit MA(1) to MA(3)
    for q in range(1, 4):
        model = ARIMA(data, order=(0, 0, q))
        fit = model.fit()
        results[f'MA({q})'] = fit
        print(f'MA({q}) AIC: {fit.aic}')

    return results
```

```python
# Run the function and store the results
results = fit_models(x)

# Determine the model with the lowest AIC
best_model = min(results, key=lambda k: results[k].aic)
print(f'The best model is: {best_model} with AIC: {results[best_model].aic}')
```

```python
# Plot the ACF and PACF for the best model
plt.figure(figsize=(12, 6))
plot_acf(x, lags=20)
plt.title('Autocorrelation Function (ACF)')
```

```python
plot_pacf(x, lags=20, method='ywm')
plt.title('Partial Autocorrelation Function (PACF)')
plt.tight_layout()
plt.show()
```
